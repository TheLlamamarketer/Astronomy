```python
def rad_to_dez(rad):
    
    dez=((rad/np.pi)*180)
    return round(dez,5)


def rad_to_deg(rad):
    Gradmaß=rad_to_dez(rad)
    Gradmaß1=np.trunc(Gradmaß)
    Gradmaß2=(Gradmaß-Gradmaß1)*60
    Gradmaß3=np.trunc(Gradmaß2)
    Gradmaß4=round(((Gradmaß2-Gradmaß3)*60),5)
    return (Gradmaß1,Gradmaß3,Gradmaß4)

def rad_to_hour(rad):
    Zeitmaß=rad_to_dez(rad)/15
    Zeitmaß1=np.trunc(Zeitmaß)
    Zeitmaß2=(Zeitmaß-Zeitmaß1)*60
    Zeitmaß3=np.trunc(Zeitmaß2)
    Zeitmaß4=round(((Zeitmaß2-Zeitmaß3)*60),5)          
    return (Zeitmaß1,Zeitmaß3,Zeitmaß4)
    

def dez_to_rad(dez):
    rad=round(((dez/180)*np.pi),5)
    return (rad)


def dez_to_deg(dez):
    degree=dez//1
    degmin=((dez%1)*60)//1
    degsec=round((((dez%1)*60)%1)*60,5)
    return(degree,degmin,degsec)

def dez_to_hour(dez):
    hour=(dez/15)//1
    min=(((dez/15)%1)*60)//1
    sec=round(((((dez/15)%1)*60)%1)*60,5)
    return(hour,min,sec)

def deg_to_rad(deg):
    degree=deg[0]
    degmin=deg[1]
    degsec=deg[2]
    rad=round((degree/180+degmin/(60*180)+degsec/(3600*180))*np.pi,5)
    return(rad)
    

def deg_to_dez(deg):
    degree=deg[0]
    degmin=deg[1]
    degsec=deg[2]
    dez=round(degree+degmin/60+degsec/3600,5)
    return(dez)


def deg_to_hour(deg):
    hour=(deg_to_dez(deg)/15)//1
    min=(((deg_to_dez(deg)/15)%1)*60)//1
    sec=round(((((deg_to_dez(deg)/15)%1)*60)%1)*60,5)
    return(hour,min,sec)


def hour_to_rad(h):
    hour=h[0]
    min=h[1]
    sec=h[2]
    rad=round((hour/12+min/(12*60)+sec/(12*3600))*np.pi,5)                   
    return(rad)


def hour_to_dez(h):
    hour=h[0]
    min=h[1]
    sec=h[2]
    dez=round((hour+min/60+sec/3600)*15,5)
    return(dez)


def hour_to_deg(h):
    hour=h[0]
    min=h[1]
    sec=h[2]
    dez=(hour+min/60+sec/3600)*15
    degree=dez//1
    degmin=((dez%1)*60)//1
    degsec=round((((dez%1)*60)%1)*60,5)
    return(degree,degmin,degsec)



    
rad=1.302029
dez=180
deg=349,33,54.9
h=7,8,0.00



print(rad,"im Bogenmaß sind",rad_to_dez(rad),"°","als Dezimalgrad")
print(rad,"im Bogenmaß sind",rad_to_deg(rad)[0],"°",rad_to_deg(rad)[1],"'",rad_to_deg(rad)[2],"''","als Gradmaß")
print(rad,"im Bogenmaß sind",rad_to_hour(rad)[0],"h",rad_to_hour(rad)[1],"m",rad_to_hour(rad)[2],"s","als Zeitmaß")
print("")
print(dez,"°","als Dezimalgrad sind",dez_to_rad(dez),"im Bogenmaß")
print(dez,"°","als Dezimalgrad sind",dez_to_deg(dez)[0],"°",dez_to_deg(dez)[1],"'",dez_to_deg(dez)[2],"''","als Gradmaß")
print(dez,"°","als Dezimalgrad sind",dez_to_hour(dez)[0],"h",dez_to_hour(dez)[1],"min",dez_to_hour(dez)[2],"sec","als Zeitmaß")
print("")
print(deg[0],"°",deg[1],"'",deg[2],"''","im Gradmaß sind",deg_to_rad(deg),"im Bogenmaß")
print(deg[0],"°",deg[1],"'",deg[2],"''","im Gradmaß sind",deg_to_dez(deg),"als Dezimalgrad")
print(deg[0],"°",deg[1],"'",deg[2],"''","im Gradmaß sind",deg_to_hour(deg)[0],"h",deg_to_hour(deg)[1],"min",deg_to_hour(deg)[2],"s","als Zeitmaß")
print("")
print(h[0],"h",h[1],"m",h[2],"sec","als Zeitmaß sind",hour_to_rad(h),"im Bogenmaß")
print(h[0],"h",h[1],"m",h[2],"sec","als Zeitmaß sind",hour_to_dez(h),"°","als Dezimalgrad")
print(h[0],"h",h[1],"m",h[2],"sec","als Zeitmaß sind",hour_to_deg(h)[0],"h",hour_to_deg(h)[1],"m",hour_to_deg(h)[2],"s","als Gradmaß")
```
